```python
import random

username = input("Please enter your username: \n").strip()

characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*<>_"

characters_set = set(characters)

loop = True

while loop:
  password = ""
  decision = input("\nWill you use a generated password? (yes/no): \n").strip().lower()

  if decision == "yes":
    count = random.randint(8,12)
    for _ in range(count):
      password += random.choice(characters)
    confirm = input(f"\nThis is your generated password: {password}\n\nDo you confirm? (yes/no): \n").strip().lower()

    if confirm == "yes":
      print(f"\nUsername: {username}\n")
      print(f"Password: {password}\n")
      loop = False

    elif confirm == "no":
      continue

    else:
      print("Please type yes or no")


  elif decision == "no":
    password = input("\nPlease type your password: \n").strip()
    password_set = set(password)
    printer = True
    while printer:
      if len(password) < 8 or len(password) > 12:
        print("\nYour password must be 8~12 characters long")
        printer = False

      elif len(password_set.difference(characters_set)) != 0:
        print("\nPlease only use letters, numbers, and !, @, #, $, %, ^, &, *, <, >, _")
        printer = False

      else:
        confirm = input(f"\nThis is your password: {password}\n\nDo you confirm? (yes/no): \n").strip().lower()

        if confirm == "no":
          printer = False
          continue

        elif confirm == "yes":
          print(f"\nUsername: {username}\n")
          print(f"Password: {password}")
          printer = False
          loop = False

        else:
          print("\nPlease type yes or no")


  else:
    print("\nPlease type yes or no")
```
